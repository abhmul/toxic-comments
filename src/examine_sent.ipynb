```python
from toxic_dataset import ToxicData, LABEL_NAMES
import os

data = "../processed_input/han/"

train_path = os.path.join(data, "train.npz")
test_path = os.path.join(data, "test.npz")
embeddings_path = os.path.join(data, "embeddings.npy")
dictionary_path = os.path.join(data, "word_to_id.pkl")

toxic = ToxicData(train_path, test_path, embeddings_path, dictionary_path)
```

```python
ids, dataset = toxic.load_train(mode="sup")
```

```python
import numpy as np

# Get the average number of sentences per doc
doc_lens = [len(sample) for sample in dataset.x]
print("MIN DOC LEN:", min(doc_lens))
print("MAX DOC LEN:", max(doc_lens))
print("AVG DOC LEN:", np.average(doc_lens))

# Get the sentence length stats
sent_lens = [len(sent) for sample in dataset.x for sent in sample]
print("MIN SENT LEN:", min(sent_lens))
print("MAX SENT LEN:", max(sent_lens))
print("AVG SENT LEN:", np.average(sent_lens))
```

```python
import matplotlib.pyplot as plt

plt.hist(doc_lens, bins='auto')
plt.xlabel("Doc Lengths")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(sent_lens, bins='auto')
plt.xlabel("Sent Lengths")
plt.ylabel("Frequency")
plt.show()
```
